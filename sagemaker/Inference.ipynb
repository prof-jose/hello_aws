# Sagemaker inference demo

This notebook illustrates real-time inference in Sagemaker. 

Pre-requisites: A trained and deployed model, such as the one generated when running KC.ipynb. 

```python
import boto3
```

```python
sm_runtime = boto3.client("runtime.sagemaker")
endpoint_name = "sagemaker-xgboost-2023-02-21-14-48-24-687"
```

```python
payload = "1180,0"

response = sm_runtime.invoke_endpoint(
    EndpointName=endpoint_name,
    ContentType="text/csv",
    Body=payload
)

response_str = response["Body"].read().decode()
```

```python
print(response_str)
```

The model is expecting data in CSV format, this is why you need to enter the `payload` variable with that specific format. 

That can look initially strange, but you should not think of endpoints like `sklearn` functions, instead as servers that get an input in a specific format, do some computations and send a response.
